```python
from scenedetect import SceneManager, open_video, ContentDetector

video_path='/content/drive/MyDrive/MetaverseAcademy/hanho/midnightJam/pyscene/github/video/898가 잘 모르는 수육 레전드 부위_1080p.mp4'
video = open_video(video_path)
scene_manager = SceneManager() #씬매니저 객체 생성
scene_manager.add_detector(ContentDetector(threshold=35.0)) #컨텐트 디텍터 설정
# threshold 27(디폴트)은 너무 자잘한 것들도 씬으로 추출됨

# Detect all scenes in video from current position to end.
scene_manager.detect_scenes(video) # 처음부터 끝까지 씬 탐지
# `get_scene_list` returns a list of start/end timecode pairs
# for each scene that was found.
scene_list = scene_manager.get_scene_list() # 탐색한 씬에 대한 시작/끝 타임코드 쌍을 리스트로 반환

for i, scene in enumerate(scene_list):
    print(
        'Scene %2d: Start %s / Frame %d, End %s / Frame %d' % (
        i+1,
        scene[0].get_timecode(), scene[0].get_frames(),
        scene[1].get_timecode(), scene[1].get_frames(),))
```
